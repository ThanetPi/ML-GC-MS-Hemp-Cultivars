```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import zscore
```

```python
dfall_10 = pd.read_excel("../GC_MS_data/hemp-gcms-threshold10.xlsx", index_col=0)
dfall_10
```

```python
dfall_10_norm = dfall_10.apply(lambda x: zscore(x), axis=1, result_type='expand')
dfall_10_norm.columns = dfall_10.columns
dfall_10_norm
```

```python
dfall_10_norm = dfall_10_norm.dropna(how='all')
dfall_10_norm
```

```python
color_map = pd.DataFrame({
    'Group': ['TH01', 'TH01', 'TH01', 'TH02', 'TH02','TH02','FS01','FS01','FS01', 'FS02', 'FS02', 'FS02']
}, index=dfall_10.columns)
```

```python
color_palette = {'TH01': 'blue', 'TH02': 'blue', 'FS01': 'lightblue', 'FS02': 'lightblue'}
```

```python
col_colors = color_map['Group'].map(color_palette)
```

```python
a=sns.clustermap(dfall_10_norm, metric='cityblock', method="average", annot=False, cmap='BuGn',
                 col_colors=col_colors,
               standard_scale=1, figsize=(6, 5), cbar_pos=((0, 0.9, .03, .1)))
bbox = a.ax_heatmap.get_position()
space = 0.0015
a.ax_heatmap.set_position([bbox.x0 + space, bbox.y0, bbox.width - space, bbox.height - space])
#plt.savefig('clustermap10p-all-hca1.png', dpi=300, bbox_inches='tight')
plt.show()
```

```python
b=sns.clustermap(dfall_10, metric='cityblock', method="average", annot=False, cmap='Blues',
               col_colors=col_colors,
               standard_scale=1, cbar_pos=((0, 0.9, .03, .1)),figsize=(6, 5))
bbox = b.ax_heatmap.get_position()
space = 0.0015
b.ax_heatmap.set_position([bbox.x0 + space, bbox.y0, bbox.width - space, bbox.height - space])
#plt.savefig('clustermap10p-all-hca2.png', dpi=300, bbox_inches='tight')
plt.show()
```
