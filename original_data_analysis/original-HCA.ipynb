```python
from scipy.stats import zscore
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
dfall= pd.read_excel("../data/hemp-gcms-ori.xlsx", index_col=0)
dfall
```

```python
dfall_norm = dfall.apply(lambda x: zscore(x), axis=1, result_type='expand')
dfall_norm.columns = dfall.columns
dfall_norm
```

```python
dfall_norm_clean = dfall_norm.dropna(how='all')
dfall_norm_clean
```

```python
color_map = pd.DataFrame({
    'Group': ['TH01', 'TH01', 'TH01', 'TH02','TH02', 'TH02', 'FS01', 'FS01', 'FS01','FS02', 'FS02', 'FS02']
}, index=dfall.columns)
```

```python
color_palette = {'TH01': 'blue', 'TH02': 'blue', 'FS01': 'lightblue', 'FS02': 'lightblue'}
```

```python
col_colors = color_map['Group'].map(color_palette)
```

```python
sns.clustermap(dfall_norm_clean, metric='cityblock', method="average", annot=False, cmap='BuGn',
               col_colors=col_colors,
               standard_scale=1, cbar_pos=((0, 0.9, .03, .1)), figsize=(12, 17))
#plt.savefig('clustermap-all-norm.png', dpi=300, bbox_inches='tight')
```

```python
sns.clustermap(dfall, metric='cityblock', method="average", annot=False, cmap='Blues',
               col_colors=col_colors,
               standard_scale=1, cbar_pos=((0, 0.9, .03, .1)), figsize=(12, 17))
#plt.savefig('clustermap-all.png', dpi=300, bbox_inches='tight')
```
